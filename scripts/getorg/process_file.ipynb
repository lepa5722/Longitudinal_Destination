```python
import csv
import pandas as pd
Headers = "CATEGORY,NAME,MANUFACTURER,PLATFORM,OS"
sample_file = "/home/hutr/iot-ipv6-project/iot-ipv6-analysis/ipv6-device-category.csv"
cur_file = "/home/hutr/iot-longitudinal/input_device_list/device_category.csv"
out_file = "/home/hutr/iot-longitudinal/input_device_list/device_information.csv"

Headers = Headers.split(",")
device_category = pd.read_csv(cur_file)
device_category_dict = {}
for index, row in device_category.iterrows():
    device_category_dict[row['Device']] = row['Category']
print(device_category_dict)
device_info = pd.read_csv(sample_file)
with open(out_file, 'w') as f2:
    writer = csv.writer(f2)
    writer.writerow(Headers)
    for device in device_category_dict.keys():
        row_sample = device_info[device_info['NAME'] == device].fillna('').values.tolist()
        if len(row_sample) > 0:
            print(row_sample)
            writer.writerow([device_category_dict[device], device, row_sample[0][3], row_sample[0][4], row_sample[0][5]])
        else:
            writer.writerow([device_category_dict[device], device, '', '', ''])

```

```python
import os
import shutil

root_dir = '/home/hutr/iot-longitudinal/output_destination/common_devices/{year}{exp}'
output_dir = '/home/hutr/iot-longitudinal/shared_files/common_devices/{year}{exp}'
years = ['2019', '2021', '2022', '2023', '2024']
exps = ['idle', 'activity', 'power']

for year in years:
    for exp in exps:
        cur_dir = root_dir.format(year=year, exp=exp)
        if not os.path.exists(cur_dir):
            continue
        # Paths to the files
        domain_file = os.path.join(cur_dir, 'domain_list/contacted_slds_with_all_dns.json')
        domain_org_file = os.path.join(cur_dir, 'domain_org/contacted_slds_with_all_dns_org.csv')
        ip_list_file = os.path.join(cur_dir, 'ip_list/all_ips.json')
        untranslated_ip_file = os.path.join(cur_dir, 'ip_list/untranslated_ips.json')
        
        # Check if all files exist
        if not (os.path.isfile(domain_file) and os.path.isfile(domain_org_file) and 
                os.path.isfile(ip_list_file) and os.path.isfile(untranslated_ip_file)):
            continue
        
        # Create the corresponding directory in the output structure
        target_dir = output_dir.format(year=year, exp=exp)
        os.makedirs(target_dir, exist_ok=True)
        
        # Copy files while maintaining structure
        shutil.copy(domain_file, os.path.join(target_dir, 'contacted_slds.json'))
        shutil.copy(domain_org_file, os.path.join(target_dir, 'contacted_slds_org.csv'))
        shutil.copy(ip_list_file, os.path.join(target_dir, 'all_ips.json'))
        shutil.copy(untranslated_ip_file, os.path.join(target_dir, 'untranslated_ips.json'))
        
        
```
